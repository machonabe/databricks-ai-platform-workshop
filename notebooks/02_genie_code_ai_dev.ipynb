
## 演習 3: Genie Code で分析クエリ生成

以下のプロンプトを試してみてください:

1. 「時間帯別の平均速度を計算してチャートにして」
2. 「バッテリー残量が 30% 以下のレコードを抽出して」
3. 「月別の走行ログ件数を集計して」

生成されたコードが正しく動作するか確認しましょう。

```python
# Genie Code が生成したコードの検証用（例: 地域別平均バッテリー）
display(spark.sql(f"""
  SELECT location, 
         ROUND(AVG(battery_pct), 1) AS avg_battery,
         COUNT(*) AS log_count
  FROM {CATALOG}.{SCHEMA}.vehicle_logs
  GROUP BY location
  ORDER BY avg_battery
"""))
```
