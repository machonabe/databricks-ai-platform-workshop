
# 共通設定 (config)

各ノートブックの先頭で `%run ./config` を実行すると、
`CATALOG` / `SCHEMA` / `VOLUME_NAME` が自動設定されます。

手動編集は不要です（既定カタログを自動検出します）。

```python
# ワークスペースの既定カタログを自動検出
CATALOG = spark.sql("SELECT current_catalog()").collect()[0][0]

# ハンズオン用スキーマ名（共有リソース用: AI Gateway 推論テーブルなど）
SCHEMA = "ai_workshop"

# ===== マルチユーザー対応: ユーザー固有スキーマ =====
# 複数ユーザーが同じワークスペースでハンズオンを実施する場合、
# テーブル書き込みは USER_SCHEMA に分離することで競合を防ぐ
CURRENT_USER = spark.sql("SELECT current_user()").collect()[0][0]
USER_SLUG = CURRENT_USER.split("@")[0].replace(".", "_").replace("-", "_")[:20]
USER_SCHEMA = f"ai_workshop_{USER_SLUG}"

# 共通 Volume 名
VOLUME_NAME = "uploads"

# ユーザー固有スキーマを自動作成
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.{USER_SCHEMA}")

print(f"=== ハンズオン共通設定 ===")
print(f"  CATALOG     : {CATALOG}")
print(f"  SCHEMA      : {SCHEMA}  ← 共有 (AI Gateway 推論テーブル等)")
print(f"  USER_SCHEMA : {USER_SCHEMA}  ← ユーザー固有テーブル書き込み先")
print(f"  VOLUME      : /Volumes/{CATALOG}/{SCHEMA}/{VOLUME_NAME}")
print(f"  USER        : {CURRENT_USER}")
```
